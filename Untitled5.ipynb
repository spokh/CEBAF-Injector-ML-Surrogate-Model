```python
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras.layers import Dense

# Create a neural network model
model = tf.keras.Sequential([
    Dense(40, activation='tanh', input_shape=(28,)),  # Input Layer
    Dense(40, activation='tanh'),                    # Layer 1
    Dense(40, activation='tanh'),                    # Layer 2
    Dense(40, activation='tanh'),                    # Layer 3
    Dense(7)                                         # Output Layer with 7 nodes for objectives
])

# Visualize the neural network architecture with custom appearance
plot_model = tf.keras.utils.plot_model(
    model,
    to_file='neural_network.png',
    show_shapes=True,
    show_layer_names=True,
    dpi=96  # Set the DPI for better quality
)

# Show the plot with custom fonts and colors
img = plt.imread('neural_network.png')

# Get the dimensions of the image
height, width, _ = img.shape

# Calculate the aspect ratio
aspect_ratio = width / height

# Set the figure size based on the image dimensions with consistent border size
border_color = 'blue'
border_width = 4
plt.figure(figsize=(width / 100, width / (100 * aspect_ratio)))  # Adjust figure size based on image dimensions

# Add a blue border around the plot
plt.gca().add_patch(plt.Rectangle((0, 0), width, height, linewidth=border_width, edgecolor=border_color, facecolor='none'))

plt.imshow(img)
plt.axis('off')

# Customize fonts and colors (you can adjust these as needed)
plt.rcParams.update({
    'text.color': 'white',  # Font color
    'font.size': 14,        # Font size
    'axes.labelcolor': 'white',
    'axes.edgecolor': 'white',
    'xtick.color': 'white',
    'ytick.color': 'white',  # Figure background color
})

# Save the plot with tight layout and without black space outside
plt.tight_layout()
plt.savefig('neuralnetwork.png', bbox_inches='tight', pad_inches=0)
plt.show()
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Dense
import matplotlib.pyplot as plt
from tensorflow.keras.utils import plot_model

# Create a neural network model
model = tf.keras.Sequential([
    Dense(40, activation='tanh', input_shape=(28,)),  # Input Layer
    Dense(40, activation='tanh'),                    # Layer 1
    Dense(40, activation='tanh'),                    # Layer 2
    Dense(40, activation='tanh'),                    # Layer 3
    Dense(7)                                         # Output Layer with 7 nodes for objectives
])

# Plot the model architecture
plot_model(model, to_file='neural_network_model.png', show_shapes=True, show_layer_names=True)

# Display the model summary
model.summary()

# Show the plot
img = plt.imread('neural_network_model.png')
plt.imshow(img)
plt.show()
```
