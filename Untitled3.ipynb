```python
import matplotlib.pyplot as plt

# Example data
f = 249.5
charge = [ 0.10, 0.20, 0.30, 0.40, 0.50, 0.60, 0.70, 0.80]  # Charge in pC (x-axis)
current = [c * f for c in charge]  # Current in µA (parallel to charge)
Bunchlength = [8, 12, 15, 11, 9, 7, 6, 4]  # Properties of matter (y-axis)

# Create a figure and axis objects
fig, ax = plt.subplots()

# Plotting the data
ax.plot(charge, Bunchlength, 'bo-')# label='Charge (pC)')

# Set the labels for x-axis and y-axis
ax.set_xlabel('Charge (pC)', fontsize=14, fontname='Times New Roman')
ax.set_ylabel('Bunch Length', fontsize=14, fontname='Times New Roman')
plt.xticks(fontsize=14, fontname='Times New Roman')
plt.yticks(fontsize=14, fontname='Times New Roman')

# Set the x-axis tick labels for both charge and current
ax.set_xticks(charge)
ax.set_xticklabels([f"{c}" for c in charge])

# Add a second x-axis for current scale
ax2 = ax.twiny()
ax2.set_xlim(ax.get_xlim())
ax2.set_xticks(charge)
plt.xticks(fontsize=14, fontname='Times New Roman')
ax2.set_xticklabels([f"{c * f:.2f}" for c in charge])
ax2.tick_params(axis='x', which='both', bottom=False, top=True)
ax2.set_xlabel('Current (µA)', fontsize=14, fontname='Times New Roman')

# Adjust the layout to create space between the charge and current scales
plt.subplots_adjust(bottom=0.25)

# Add a legend to the plot
#ax.legend()
#plt.title('hello')
# Display the plot
plt.show()
#plt.xlabel('X-axis label', fontsize=10, )
#plt.ylabel('Y-axis label', fontsize=10)
#plt.title('Plot Title', fontsize=12)
#plt.xticks(fontsize=8, fontname='Times New Roman')
#plt.yticks(fontsize=8, fontname='Times New Roman')
```
