```python
import pandas as pd
from scipy.integrate import quad
import numpy as np

# Function to integrate: sqrt(delta f / f) * cos(2pi f z/c)
def integrand(z, delta_f_over_f, frequency, c):
    omega = 2 * np.pi * frequency
    return np.sqrt(delta_f_over_f) * np.cos(omega * z / c)

# Read data from Excel file  # Assuming the Excel file is in the same directory
df = pd.read_excel('Beadpull.xlsx')

# Extract z, delta_f_over_f, frequency, and c (replace column names with your actual column names)
z_values = df['z (m)'].values
delta_f_over_f_values = df['delta-f']/ df['fo (MHz)'].values  # Replace with your actual column name
frequency_values = df['fo (MHz)'].values  # Replace with your actual column name
c_value = 3e8  # Replace with the speed of light or your specific value

# Set integration limits (z values from 0 m to 0.1 m)
z_start = 0.02 # 0 m
z_end = 0.24 # 0.1 m

# Perform integration for each (z, delta_f_over_f, frequency, c) quadruple
integrals = []
for z, delta_f_over_f, frequency in zip(z_values, delta_f_over_f_values, frequency_values):
    result, _ = quad(integrand, z_start, z_end, args=(delta_f_over_f, frequency, c_value))
    integrals.append(result)

# Add the integration results to the DataFrame
df['integral_result'] = integrals

# Calculate the total integral value
total_integral = sum(integrals)

# Print the total integral value
print(f'Total Integral Value: {total_integral}')
```

```python
import numpy as np

# Given values
alpha_c = 1e-3  # Momentum compaction factor
f0 = 3e9  # Revolution frequency in Hz

# Function to calculate amplitude of phase oscillation
def calculate_phase_amplitude(frequency):
    nu_s = frequency / f0
    theta_s = np.abs(alpha_c * nu_s)
    return theta_s

# Calculate amplitudes for 500 kHz and 3 GHz
amplitude_500kHz = calculate_phase_amplitude(5e5)  # 500 kHz in Hz
amplitude_3GHz = calculate_phase_amplitude(3e9)  # 3 GHz in Hz

print(f"Amplitude at 500 kHz: {amplitude_500kHz} radians")
print(f"Amplitude at 3 GHz: {amplitude_3GHz} radians")
```
