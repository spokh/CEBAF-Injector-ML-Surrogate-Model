## Chopper Scan at 98 $\mu$A

```python
import pandas as pd
data = pd.read_csv('100muAPcup_98uAgunPBoff.csv',  delim_whitespace=True) 
# Replace 'data.txt' with the path to your .txt file
data.head()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x= data["!TimeStamp"]
y=data["Input:choppergang"]
z=data["Output:Fcupcorrected"]
# Create longitudinal current profile plot
plt.plot(x-np.mean(x), z,'.-r')
plt.xlabel("t (ps)", fontsize=12)
plt.ylim(0)
plt.title("Exp.longitudinal profile at 98 $\mu$A")
plt.ylabel("Current ($\mu$ A)",fontsize=14)
plt.show()
```
