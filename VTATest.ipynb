```python
gradient = CW_data['Eacc']
freq = CW_data['FreqHz']
square_gradient = gradient ** 2
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
# Plot 'x' against 'y'
plt.plot(square_gradient, freq, marker='o', linestyle='-', color='r', label='Data')

# Add labels and legend
plt.xlabel('$E_{acc}^2$ $(MV/m)^2$')
plt.ylabel('Cavity frequency (Hz)')
plt.title('Cavity frequecny as a function of accelerating gradient square')
plt.legend()

# Show the plot
plt.show()
```

```python
# Define a linear function for the curve fit
def linear_function(x, a, b):
    return a * x + b

# Perform the curve fit
params, covariance = curve_fit(linear_function,square_gradient, freq)

# Get the fitted parameters
a_fit, b_fit = params

# The slope (a) is the first parameter in the 'params' array
slope = a_fit

# Print the fitted parameters including the slope
print("Fitted parameters:")
print("a (slope) =", slope)
print("b =", b_fit)
```

$$E_{acc}(V/m) =\sqrt{{4\pi f_0\tau(\mathrm{P_{f}}}+ C_\beta \sqrt{\mathrm{P_{r}P_{f}}})\frac{(\frac{r}{Q})}{L}}$$
Diffrentiating above with respect to $\tau$

$$(\frac{\partial E}{\partial\tau})^2= {4\pi f_0(\mathrm{P_{f}}}+ C_\beta \sqrt{\mathrm{P_{r}P_{f}}})\frac{(\frac{r}{Q})}{L} \frac{1}{4\tau}$$
Diffrentiating  with respect to $P_f$
$$(\frac{\partial E}{\partial P_f})^2= {4\pi f_0 \tau} \frac{(\frac{r}{Q})}{L} (1+ \frac{C_\beta\sqrt{P_r}}{2\sqrt{P_f}})^2\frac{1}{4(\mathrm{P_{f}}+ C_\beta \sqrt{\mathrm{P_{f}}\mathrm{P_{r}}}})$$
Diffrentiating  with respect to $P_r$
$$(\frac{\partial E}{\partial P_r})^2= {4\pi f_0 \tau} \frac{(\frac{r}{Q})}{L} \frac{C_\beta^2P_f}{4P_r(\mathrm{P_{f}}+ C_\beta \sqrt{\mathrm{P_{f}}\mathrm{P_{r}}}})$$

Now, 
$$\Delta E = \sqrt{(\frac{\partial E}{\partial\tau})^2 (\Delta\tau)^2 + (\frac{\partial E}{\partial P_f})^2(\Delta P_f)^2 + (\frac{\partial E}{\partial P_r})^2 (\Delta P_r)^2}$$
Substituting the values, we get the error in the gradient as 
$$\Delta E = \frac{E}{2} \sqrt{\left( \frac{(2P_f + C_\beta \sqrt{P_f P_r}) \Delta P_f} {2(P_f + C_\beta \sqrt{P_f P_r})P_f } \right)^2  + \left( \frac{ \sqrt{P_r}) \Delta P_r} {2(P_f + C_\beta \sqrt{P_r})P_r } \right)^2 +
\left( \frac{\Delta \tau}{\tau}\right)^2}$$
