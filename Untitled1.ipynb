```python
# Sort the data frame by one of the columns in descending order
df1 = df.sort_values(by='uf:nemixrms', ascending=False)

# Create an empty list to store the Pareto front
pareto_front = []

# Iterate through the sorted data frame, adding each row to the Pareto front if it is not dominated by any of the previous rows in the front
for index, row in df1.iterrows():
    if not any((row['uf:nemixrms'] <x['uf:nemixrms']) and (row['uf:stdt'] < x['uf:stdt']) for x in pareto_front):
        pareto_front.append(row)

# Convert the Pareto front to a data frame
pareto_df = pd.DataFrame(pareto_front)
#plt.scatter(stdt*1e+13, stdx*1000, color='blue', label='Data Points')
# Plot the Pareto front using a scatter plot
plt.plot(pareto_df['uf:stdt'], pareto_df['uf:nemixrms'])
plt.xlabel('uf:')
plt.ylabel('uf:stdt')
plt.show()
```
