```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data into a DataFrame
#data = pd.read_csv('data.csv')  # Replace 'data.csv' with the path to your data file

# Plot two columns
nemix = data['uf:nemixrms']  # Replace 'column1' with the name of the first column
nemiy = data['uf:nemiyrms']# Replace 'column2' with the name of the second column
stdt = data ['uf:stdt']
stdx = data['uf:stdx']
stdy = data['uf:stdy']
stdek =data['uf:stdEk_eV']           
plt.plot(stdx, stdt)
plt.xlabel('std(x)[m]')
plt.ylabel('std(t) [s]')
plt.title('Plot of std(x)  vs std(t)')
plt.show()
```

```python
import matplotlib.pyplot as plt
# Plot the data points
plt.scatter(stdx, stdt, color='blue', label='Data Points')

# Identify the Pareto front
pareto_front = []
for i in range(len(stdx)):
    is_pareto = True
    for j in range(len(stdx)):
        if stdx[j] >= stdx[i] and stdt[j] >= stdt[i] and j != i:
            is_pareto = False
            break
    if is_pareto:
        pareto_front.append((stdx[i], stdt[i]))

# Extract the x and y values for the Pareto front
pareto_x = [point[0] for point in pareto_front]
pareto_y = [point[1] for point in pareto_front]

# Plot the Pareto front
plt.plot(pareto_x, pareto_y, color='red', label='Pareto Front')

# Set labels and title for the plot
plt.xlabel('stdx')
plt.ylabel('stdt')
plt.title('Pareto Front')

# Add legend
plt.legend()

# Display the plot
plt.show()
```

```python
input_file = 'mgoresults.txt'  # Replace with the path to your large data file in .txt format
output_file = 'new_data.txt'  # Replace with the desired file path and name for the new data file

# Select the desired columns from the large dataset
selected_columns = [2, 3, 4]  # Replace with the indices of the columns you want to select (starting from 0)

# Read the large data file and extract the selected columns
with open(input_file, 'r') as file:
    lines = file.readlines()

selected_data = []
for line in lines:
    values = line.strip().split()
    selected_values = [values[col_index] for col_index in selected_columns]
    selected_data.append(selected_values)

# Save the selected data to a new file
with open(output_file, 'w') as file:
    for selected_values in selected_data:
        line = '\t'.join(selected_values) + '\n'  # Assuming tab-separated values, modify the delimiter if needed
        file.write(line)
```
