```python
import matplotlib.pyplot as plt
# Plot the data points
plt.scatter(stdx, stdt, color='blue', label='Data Points')

# Identify the Pareto front
pareto_front = []
for i in range(len(stdx)):
    is_pareto = True
    for j in range(len(stdx)):
        if stdx[j] >= stdx[i] and stdt[j] >= stdt[i] and j != i:
            is_pareto = False
            break
    if is_pareto:
        pareto_front.append((stdx[i], stdt[i]))

# Extract the x and y values for the Pareto front
pareto_x = [point[0] for point in pareto_front]
pareto_y = [point[1] for point in pareto_front]

# Plot the Pareto front
plt.plot(pareto_x, pareto_y, color='red', label='Pareto Front')

# Set labels and title for the plot
plt.xlabel('stdx')
plt.ylabel('stdt')
plt.title('Pareto Front')

# Add legend
plt.legend()

# Display the plot
plt.show()
```
